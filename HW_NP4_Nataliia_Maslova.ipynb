5. Стандартизуйте дані:

Застосуйте процес стандартизації даних за допомогою функцій з конспекту або бібліотеки sklearn.

```python
def data_scale(data, scaler_type='minmax'):
    """
    Scale the data using the specified scaler.

    :param data: The data to be scaled, as a NumPy array or a pandas DataFrame.
    :param scaler_type: The type of scaler to use. Options are 'minmax', 'std', or 'norm'.
    :return: Scaled data as a NumPy array.
    """
    from sklearn.preprocessing import MinMaxScaler, StandardScaler, Normalizer

    # Dictionary mapping scaler types to corresponding scaler objects
    scalers = {
        'minmax': MinMaxScaler(),
        'std': StandardScaler(),
        'norm': Normalizer()
    }

    # Get the scaler based on the scaler_type
    scaler = scalers.get(scaler_type)

    if scaler is None:
        raise ValueError(f"Invalid scaler_type '{scaler_type}'. Options are: {', '.join(scalers.keys())}")

    # Fit and transform the data
    scaler.fit(data)
    result = scaler.transform(data)

    return result
```

```python
data_scaled = data_scale(df, 'std')
df_scaled = pd.DataFrame(data_scaled, columns=[df.columns])
df_scaled.head()
```

7. Обчисліть матриці відстаней:

Використовуйте алгоритми та функції з конспекту для обчислення матриці відстаней для різних метрик: cityblock, cosine, euclidean, l1, manhattan.

```python
from sklearn.metrics.pairwise import pairwise_distances

def distance_matrix(df, metric='euclidean'):
    """
    Обчислює матрицю відстаней для цього DataFrame.

    Параметри:
    - df: pandas DataFrame
      Вхідні дані.
    - metric: str, опціонально (за замовчуванням 'euclidean')
      Метрика відстані, яку ви хочете використовувати.

    Повертає:
    - pandas DataFrame
      Матриця відстаней.
    """
    numeric_columns = df.select_dtypes(include='number')

    # Обчислити матрицю відстаней
    distance_matrix = pairwise_distances(numeric_columns, metric=metric)

    # Створити DataFrame з результатами
    result_df = pd.DataFrame(distance_matrix, index=numeric_columns.index, columns=numeric_columns.index)

    return result_df
```

8. Візуалізуйте отримані матриці:

Використайте функцію heatmap з бібліотеки seaborn або інші методи візуалізації для представлення отриманих матриць відстаней.

```python
distances = ['euclidean', 'l1', 'cityblock', 'manhattan', 'l2', 'cosine']
for dist in distances:
    plt.figure(figsize=(10, 8))
    distance_matrix = pairwise_distances(df_bc.sort_values(by='cancer'), metric=dist)
    # To get the lower triangular matrix, you can use:
    mask_lower = np.tril(np.ones_like(distance_matrix, dtype=bool))
    lower_triangle_matrix = np.where(mask_lower, distance_matrix, np.nan)

    # Apply mask to get upper triangular matrix
    # upper_triangle_matrix = np.where(mask_upper, distance_matrix, np.nan)

    plt.imshow(lower_triangle_matrix, cmap="vlag", interpolation="nearest")
    # Налаштування заголовку та підписів осей
    plt.title(f'{dist} матриця подібностей', fontsize=16)
    plt.xlabel('Зразки', fontsize=14)
    plt.ylabel('Зразки', fontsize=14)
    plt.show()
```
