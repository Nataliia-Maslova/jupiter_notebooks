Запиши систему обмежень, виходячи з умов. Не забудь врахувати час роботи спеціалістів за кожним видом робіт, місячне навантаження спеціалістів та неможливість від’ємних значень.

Знайди оптимальний розподіл послуг за допомогою бібліотеки SciPy так, як це робилось у конспекті.

Знайди сумарний місячний дохід компанії при оптимальному розподілі функцій.

Зроби висновки.

```python
from scipy.optimize import linprog

# Коефіцієнти цільової функції (для мінімізації)
c = [-2, -9, -6]

# Коефіцієнти для нерівностей
A_ub = [
    [12, 6, 2],   # Дизайнери
    [12, 24, 18], # Програмісти
    [12, 18, 12]  # Тестувальники
]

# Праві частини нерівностей
b_ub = [320, 192, 180]

x_bounds = (0, None)

result = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=[x_bounds, x_bounds, x_bounds], method='highs')

print(result)
```

Інтерпретація результатів
Оптимальний прибуток: Компанія може максимізувати прибуток до 72 тис. доларів на місяць, виконуючи 8 інтернет-магазинів і не виконуючи інших робіт.

Час роботи спеціалістів:

Дизайнери використовують 48 годин із доступних 320 годин.

Програмісти повністю завантажені (192 години).

Тестувальник використовує 144 години із доступних 180 годин.


Це означає, що оптимальним рішенням є зосередження на розробці інтернет-магазинів, оскільки це приносить найбільший прибуток у межах доступного часу роботи спеціалістів.
