Аналітично виконай диференціювання функції та знайди точку максимуму. Програмно візуалізуй функцію, її похідну та точку максимуму.

$$
y=-3x^2+30x
$$
$$
f(x)' = (-3x^2+30x)' = -6x + 30
$$
\
точка максимуму\
−6x+30=0\
x = 5
$$
y = -3*5^2 +30*5 = 75
$$
Точка максимуму Max(5,75)

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.integrate import quad
```

```python
fig, ax = plt.subplots()
# Перемістимо лівий і нижній стовпчики до x = 0 і y = 0 відповідно.
ax.spines[["left", "bottom"]].set_position(("data", 0))

# Сховати верхню та праву лінію
ax.spines[["top", "right"]].set_visible(False)

# Намалюємо стрілки (як чорні трикутники: ">k"/"^k") на кінцях осей.
# Також вимкнемо відсікання (clip_on=False) стрілок.
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

# Додамо проміжні лінії
ax.grid(True, linestyle='-.')

# x, y = sp.symbols('x y')
x = sp.Symbol('x')

f = -3*x**2 + 30*x
df = f.diff()

x1 = np.linspace(-6, 6, 500)
f_lambd = sp.lambdify(x, f, 'numpy')
# y = [f.subs(x, i) for i in x1]
y = f_lambd(x1)

df_lambd = sp.lambdify(x, df, 'numpy')
dy = df_lambd(x1)

x_max = np.array([float(sp.expand_complex(i)) for i in sp.solve(df, x)])
y_max = f_lambd(x_max)

sns.lineplot(x=x1, y=y, ax=ax)
sns.lineplot(x=x1, y=dy, ax=ax)
sns.scatterplot(x=x_max, y=y_max, ax=ax, color='red')
```
