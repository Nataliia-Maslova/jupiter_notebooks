1. Завантажити набір даних за допомогою load_iris() бібліотеки sklearn.

```python
from sklearn.datasets import load_iris

# Завантаження даних Iris
iris_data = load_iris()

# Отримання ознак та міток класів
features = iris_data.data
labels = iris_data.target

# Виведення перших декількох рядків даних
print("Перші п'ять рядків даних:")
print(features[:5])
print("Мітки класів:")
print(labels[:5])
```

7. Реалізувати функцію обчислення значень дискримінантної функції для одного рядка (вектора) тестових даних.

```python
def g(x_test, X_class, X_cov, X_inv_cov, X_prob):
    """Повертає значення дискримінантної функції для вектора даних

    Parameters
    ----------
    x_test : вектор тестових даних
             Array of samples/test vectors.
    X_class:

    Returns
    -------
    C : ndarray of shape (n_samples, n_classes)
        Posterior log-probabilities of classification per class.
    """
    cov_det = np.linalg.det(X_cov)
    log_cov_det = np.log(cov_det)
    x_mean = np.reshape(np.mean(X_class, axis=0), (np.shape(X_class)[1],1))
    diff = x_test.reshape(-1,1) - x_mean
    dif_cov = np.dot(diff.transpose(), X_inv_cov)
    diff_res = np.dot(dif_cov, diff)

    g_i = -1/2*diff_res - 1/2*log_cov_det+np.log(X_prob)
    return g_i
```

```python
def g_matrix(X_test, X_class, X_cov, X_inv_cov, X_prob):

    g_val = []
    g_proba = []

    for i in range(len(X_test)):
        x_test = X_test[i]
        g_0 = g(x_test, X_class[0], X_cov[0], X_inv_cov[0], X_prob[0])
        g_1 = g(x_test, X_class[1], X_cov[1], X_inv_cov[1], X_prob[1])
        g_2 = g(x_test, X_class[2], X_cov[2], X_inv_cov[2], X_prob[2])
        g_i = np.array(list([g_0[0][0], g_1[0][0], g_2[0][0]]))
        values = g_i.reshape(1,-1)
        likelihood = np.exp(values - values.max(axis=1)[:, np.newaxis])
        res = likelihood / likelihood.sum(axis=1)[:, np.newaxis]

        g_val.append(values)
        g_proba.append(res)

    return np.array(g_val), np.array(g_proba)
```

8. Реалізувати функцію обчислення значень дискримінантної функції та імовірностей приналежності кожному класу для всієї матриці тестових даних.

```python
g_val, g_proba = g_matrix(features_test, [x_0, x_1, x_2], [x_0_cov, x_1_cov, x_2_cov], [inv_cov_0,inv_cov_1,inv_cov_2], [P_0, P_1, P_2])
print('Значення дискримінантної функції\n', g_val)

print('Значення імовірностей класів\n', g_proba)
```

9. Виконати прогнозування на тестових даних за допомогою функії QuadraticDiscriminantAnalysis() бібліотеки sklearn та порівняти отримані результати.

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

features_train, features_test, labels_train, labels_test

qda = QuadraticDiscriminantAnalysis()
qda.fit(features_train, labels_train)

values_qda = qda._decision_function(features_test)

proba_qda = qda.predict_proba(features_test)

print('Значення дискримінантної функції\n', values_qda)

print('Значення імовірностей класів\n', proba_qda)
```

```python
own = [np.argmax(i) for i in g_proba]
own
```

```python
sk = [np.argmax(i) for i in proba_qda]
sk
```

```python
def compare_lists(list1, list2):
    # Convert lists to sets to remove duplicates and enable set operations
    set1 = set(list1)
    set2 = set(list2)

    # Find the intersection of both sets (i.e., common elements)
    common_elements = set1.intersection(set2)

    # Calculate the number of common elements
    num_common = len(common_elements)

    # Calculate the total number of unique elements in both lists
    total_unique = len(set1.union(set2))

    # Calculate the percentage of matching elements
    if total_unique == 0:
        percentage = 0.0
    else:
        percentage = (num_common / total_unique) * 100

    return percentage

compare_lists(own, sk)
```
