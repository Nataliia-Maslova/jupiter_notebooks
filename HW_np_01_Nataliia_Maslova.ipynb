1. Завантажити та створити DataFrame
2. Отримати базові статистичні характеристики:

```python
import numpy as np
from sklearn.datasets import load_iris
iris_data = load_iris()
print(iris_data.DESCR)
```

```python
import pandas as pd
iris_data = load_iris()
iris_df = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
print(iris_df.describe())
```

3. Візуалізувати розподілення спостережень за класами:

```python
import seaborn as sns
iris_df['target'] = iris_data.target
selected_columns = ['sepal length (cm)', 'sepal width (cm)', 'target']
sns.pairplot(iris_df[selected_columns], hue='target', palette='tab10', diag_kind='hist')
```

```python
selected_columns = ['petal length (cm)', 'petal width (cm)', 'target']
sns.pairplot(iris_df[selected_columns], hue='target', palette='tab10', diag_kind='hist')
```

```python
selected_columns = ['sepal length (cm)', 'petal length (cm)', 'target']
sns.pairplot(iris_df[selected_columns], hue='target', palette='tab10', diag_kind='hist')
```

```python
selected_columns = ['sepal width (cm)', 'petal width (cm)', 'target']
sns.pairplot(iris_df[selected_columns], hue='target', palette='tab10', diag_kind='hist')
```

4. Виконати стандартизацію даних

```python
features = iris_df.iloc[:, :-1]
features.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled = scaler.fit_transform(features)
```

```python
df_scaled = pd.DataFrame(scaled, columns=features.columns)
df_scaled.head()
```

5. Виконати спектральну кластеризацію:

```python
from sklearn.cluster import SpectralClustering
n_clusters = 3
spectral_clustering = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42)
clusters = spectral_clustering.fit_predict(df_scaled)
df_scaled['cluster'] = clusters
df_scaled.head()
```

6. Порівняти спрогнозовані кластери та дійсні класи:

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

true_labels = iris_data.target
conf_matrix = confusion_matrix(true_labels, clusters)
print(conf_matrix)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Cluster 1', 'Cluster 2', 'Cluster 3'],
            yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.xlabel('Predicted Clusters')
plt.ylabel('True Classes')
plt.title('Confusion Matrix between True Classes and Predicted Clusters')
plt.show()
```

7. Візуалізувати результати кластеризації:

```python

x_feature = 'sepal length (cm)'
y_feature = 'sepal width (cm)'


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_scaled, x=x_feature, y=y_feature, hue='cluster', palette='tab10', s=100)


plt.title('Spectral Clustering Results on Iris Dataset')
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.legend(title='Cluster')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris_df, x=x_feature, y=y_feature, hue=iris_data.target, palette='tab10', s=100)


plt.title('True Classes of Iris Dataset')
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.legend(title='True Class')
plt.show()
```
